# Criar Dataframe para Análise Vertical para deteminada empresa

```python
# Filtrar empresa por CNPJ
bpa_seer = bpa[ bpa['CNPJ_CIA'] == '04.986.320/0001-13']

# Filtrar apenas as colunas que serão utilizadas
bpa_seer = bpa_seer[['CD_CONTA','DS_CONTA','VL_CONTA']]

# Converter valores numéricos
bpa_seer['VL_CONTA'] = pd.to_numeric(bpa_seer['VL_CONTA'])

# Reduzir as contas
bpa_seer = bpa_seer[bpa_seer['CD_CONTA'].str.len() <= 7]

# Reiniciar os indices
bpa_seer.reset_index(inplace=True, drop=True)

# Obter o Ativo Total
ativo = bpa_seer.loc[0]['VL_CONTA']

# Criar coluna (%)
bpa_seer['%'] = bpa_seer['VL_CONTA'] / ativo * 100

bpa_seer
```

# Criar DataFrame para análises verticais para 3 anos - 2021/2022/2023


## Utilizando Concat

```python
bpa_seer_varios_anos = pd.DataFrame()

for ano in range(2021,2024):
  # # Ler informações de arquivo (.csv) específico
  bpa = pd.read_csv(pasta_drive+f'/dfp/dfp_cia_aberta_BPA_con_{ano}.csv', sep=';', encoding='ISO-8859-9', decimal=',')


  # # Filtrar informações da coluna ORDEM_EXERC == ÚLTIMO
  bpa = bpa[bpa['ORDEM_EXERC'] == 'ÚLTIMO']

  # # Filtrar empresa por CNPJ
  bpa_seer = bpa[ bpa['CNPJ_CIA'] == '04.986.320/0001-13']

  # # Filtrar apenas as colunas que serão utilizadas
  bpa_seer = bpa_seer[['DT_REFER','CD_CONTA','DS_CONTA','VL_CONTA']]

  # # Converter valores numéricos
  bpa_seer['VL_CONTA'] = pd.to_numeric(bpa_seer['VL_CONTA'])

  # # Reduzir as contas
  bpa_seer = bpa_seer[bpa_seer['CD_CONTA'].str.len() <= 7]

  # # Reiniciar os indices
  bpa_seer.reset_index(inplace=True, drop=True)

  # # Obter o Ativo Total
  ativo = bpa_seer.loc[0]['VL_CONTA']

  # # Criar coluna (%)
  # bpa_seer['%'] = bpa_seer['VL_CONTA'] / ativo * 100

  # transformação dos indices, colunas e linhas
  bpa_seer = bpa_seer.pivot(index=['CD_CONTA','DS_CONTA'], columns='DT_REFER', values='VL_CONTA')

  # criar (%) com iloc (valores de indices das colunas)
  bpa_seer['%'] = bpa_seer.iloc[:,0:] / ativo * 100

  # Agrupamento por colunas
  bpa_seer_varios_anos = pd.concat([bpa_seer_varios_anos, bpa_seer], axis=1)
  # display(bpa_seer.head())
  # break
bpa_seer_varios_anos
```

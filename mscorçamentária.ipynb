```python
import requests, json


```

```python
class Gov:

    def MSCOrcamentaria(id_ente, an_referencia):
        id_tv = ['beginning_balance','ending_balance','period_change']
        me_referencia = 6
        classe_conta = [5,6]
        co_tipo_matriz = ['MSCC','MSCE']

        ResultMscOrc = [[],[],[],[],[],[],[],[],[],[],[],[]] # resultado de cada laço
        contador = 0

        for i in co_tipo_matriz:
            for j in id_tv:
                for k in classe_conta:

                    response = requests.get(f'https://apidatalake.tesouro.gov.br/ords/siconfi/tt/msc_orcamentaria?id_ente={id_ente}&an_referencia={an_referencia}&me_referencia={me_referencia}&co_tipo_matriz={i}&classe_conta={k}&id_tv={j}')
                    if(response.status_code != 200):
                        ResultMscOrc[contador] = {result:'Falhou'}
                    else:
                        ResultMscOrc[contador] = json.loads(response.text)['items']
                    contador+=1

        return ResultMscOrc
    def Anexo():# Anexos("cidades") do Brasil.
        response = requests.get('http://apidatalake.tesouro.gov.br/ords/siconfi/tt/entes')
        if(response.status_code != 200):
           return({result:'Falhou'})
        return json.loads(response.text)['items']
```

```python
Anexos = Gov.Anexo()
Cidade = [['Florianópolis'],[]]
for i in Anexos:
    if(i['ente'] in Cidade[0]):
        Cidade[1]+=[i]

print(Cidade[1])



```

```python
# -*- coding: utf-8 -*-
#importes do modulo
import requests, json
```

```python
class Ente:
    def Anexo():# construção de catalogar todos os Anexos("cidades") do Brasil.
        response = requests.get('http://apidatalake.tesouro.gov.br/ords/siconfi/tt/entes')
        if(response.status_code != 200):
            return({result:'Falhou'})
        return json.loads(response.text)['items']

    def DCA(an_exercicio, no_anexo, id_ente): #Pegar por ano o DCA daquele ente
        response = requests.get(f'https://apidatalake.tesouro.gov.br/ords/siconfi/tt/dca?an_exercicio={an_exercicio}&no_anexo={no_anexo}&id_ente={id_ente}')
        if(response.status_code != 200):
            return({result:'Falhou'})
        return json.loads(response.text)['items']

    def MSCPatrimonial(id_ente, an_referencia):
        id_tv = ['beginning_balance','ending_balance','period_change']
        me_referencia = 6
        classe_conta = [1,2,3,4]
        co_tipo_matriz = ['MSCC','MSCE']

        ResultMscPat = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]] # resultado de cada laço
        contador = 0

        for i in co_tipo_matriz:
            for j in id_tv:
                for k in classe_conta:

                    response = requests.get(f'https://apidatalake.tesouro.gov.br/ords/siconfi/tt/msc_patrimonial?id_ente={id_ente}&an_referencia={an_referencia}&me_referencia={me_referencia}&co_tipo_matriz={i}&classe_conta={k}&id_tv={j}')
                    if(response.status_code != 200):
                        ResultMscPat[contador] = {result:'Falhou'}
                    else:
                        ResultMscPat[contador] = json.loads(response.text)['items']

                    contador+=1

        return ResultMscPat

    def MSCOrcamentaria(id_ente, an_referencia):
        id_tv = ['beginning_balance','ending_balance','period_change']
        me_referencia = 6
        classe_conta = [5,6]
        co_tipo_matriz = ['MSCC','MSCE']

        ResultMscOrc = [[],[],[],[],[],[],[],[],[],[],[],[]] # resultado de cada laço
        contador = 0

        for i in co_tipo_matriz:
            for j in id_tv:
                for k in classe_conta:

                    response = requests.get(f'https://apidatalake.tesouro.gov.br/ords/siconfi/tt/msc_orcamentaria?id_ente={id_ente}&an_referencia={an_referencia}&me_referencia={me_referencia}&co_tipo_matriz={i}&classe_conta={k}&id_tv={j}')
                    if(response.status_code != 200):
                        ResultMscOrc[contador] = {result:'Falhou'}
                    else:
                        ResultMscOrc[contador] = json.loads(response.text)['items']
                    contador+=1

        return ResultMscOrc

    def MSCControle(id_ente, an_referencia):
        id_tv = ['beginning_balance','ending_balance','period_change']
        me_referencia = 6
        classe_conta = [7,8]
        co_tipo_matriz = ['MSCC','MSCE']

        ResultMscCon = [[],[],[],[],[],[],[],[],[],[],[],[]] # resultado de cada laço
        contador = 0

        for i in co_tipo_matriz:
            for j in id_tv:
                for k in classe_conta:

                    response = requests.get(f'https://apidatalake.tesouro.gov.br/ords/siconfi/tt/msc_controle?id_ente={id_ente}&an_referencia={an_referencia}&me_referencia={me_referencia}&co_tipo_matriz={i}&classe_conta={k}&id_tv={j}')
                    if(response.status_code != 200):
                        ResultMscCon[contador] = {result:'Falhou'}
                    else:
                        ResultMscCon[contador] = json.loads(response.text)['items']

                    contador+=1

        return ResultMscCon
```
